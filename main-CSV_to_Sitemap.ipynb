```python
# prompt: pandas membagi menjadi per 100 data csv dijadikan txt, ambil colom address-nya saja. tambahkan agar outputnya menjadi format sitemap

import pandas as pd
import os

def process_csv_to_sitemap(csv_filepath, output_dir):
    """
    Processes a CSV file, splits it into chunks of 100 rows,
    extracts the 'address' column, and formats the output as a sitemap.
    """

    try:
        df = pd.read_csv(csv_filepath, delimiter=';')  # Adjust if your file uses another delimiter
    except FileNotFoundError:
        print(f"Error: File not found at {csv_filepath}")
        return
    except pd.errors.EmptyDataError:
        print(f"Error: Empty CSV file at {csv_filepath}")
        return
    except pd.errors.ParserError:
        print(f"Error: Could not parse CSV file at {csv_filepath}")
        return

    if 'Address' not in df.columns:
        print(f"Error: 'Address' column not found in the CSV file.")
        return

    os.makedirs(output_dir, exist_ok=True)

    chunk_size = 100
    for i in range(0, len(df), chunk_size):
        chunk = df.iloc[i:i + chunk_size]
        sitemap_content = """<?xml version="1.0" encoding="UTF-8"?>
<urlset xmlns="http://www.sitemaps.org/schemas/sitemap/0.9">
"""
        for address in chunk['Address']:
            sitemap_content += f"""    <url>
        <loc>{address}</loc>
        <lastmod>2024-08-01T16:51:13+00:00</lastmod>
    </url>
"""

        sitemap_content += """</urlset>"""
        output_filename = os.path.join(output_dir, f"sitemap_{i // chunk_size}.xml")
        with open(output_filename, "w") as f:
            f.write(sitemap_content)
        print(f"Sitemap chunk {i//chunk_size} created successfully at: {output_filename}")

# Example usage
csv_file_path = '/content/internal_all_baru.csv'  # Replace with your CSV file path
output_directory = '/content/drive/MyDrive/WP-Posting' # Replace with the desired output directory
process_csv_to_sitemap(csv_file_path, output_directory)
```
